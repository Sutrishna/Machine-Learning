```python
from sklearn.covariance import MinCovDet

from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(data)
```

```python
cov_robust = MinCovDet()
```

```python
cov_robust.fit(data)
```

```python
sigma = cov_robust.covariance_
```

```python
import scipy
sigma = scipy.linalg.inv(sigma)
```

```python
from scipy.spatial.distance import mahalanobis
```

```python
mean = data.mean()
```

```python
mean
```

```python
data = pd.DataFrame(data)
```

```python
data.shape
```

```python
range(data.shape[0])
```

```python
X= []
for i in range(data.shape[0]):
    x= float(mahalanobis(data.loc[i,:], mean ,sigma) ** 2)
    X.append(x)
```

```python
plt.figure(figsize=(100, 30))

p1 = [0, 50] 
p2 = [400, 50]

plt.plot(range(data.shape[0]), X)
plt.plot(p1, p2)
plt.show()
```

```python
index_to_drop = []

for i in range(len(X)):
    if X[i] > 50.0:
        index_to_drop.append(i)
```

```python
index_to_drop
```

```python
data = data.drop(data.index[[index_to_drop]])
```
